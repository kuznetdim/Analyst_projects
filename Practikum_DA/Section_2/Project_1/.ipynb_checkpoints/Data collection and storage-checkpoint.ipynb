**<font color='blue'>Проверьте гипотезу: «Средний спрос на билеты во время фестивалей не отличается от среднего спроса на билеты в обычное время».**</font>

**<font color='blue'>нулевая гипотеза:</font>  Средний спрос на билеты во время фестивалей не отличается от среднего спроса на билеты в обычное время, т е количества покупаемых билетов равны  
<font color='blue'>альтернативная гипотеза:</font> Средний спрос на билеты во время фестивалей отличается от среднего спроса на билеты в обычное время, т е количества покупаемых билетов не равны**

```python
query_last['festival_name'] = query_last['festival_name'].fillna(value = 0)
```

```python
query_last
```

```python
fest_ticket = query_last.query('festival_name != 0')
```

```python
fest_ticket
```

```python
no_fest_ticket = query_last.query('festival_name == 0')
```

```python
no_fest_ticket
```

```python
fest_ticket['ticket_amount'].describe()
```

```python
no_fest_ticket['ticket_amount'].describe()
```

**<font color='blue'>Cтандартное отклонение отличается более чем в 2 раза - ставлю False у `equal_var`**</font>

```python
alpha = .05
```

```python
results = st.ttest_ind(
    fest_ticket['ticket_amount'], 
    no_fest_ticket['ticket_amount'],
    equal_var = False)
```

```python
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод: Средний спрос на билеты во время фестивалей не отличается от среднего спроса на билеты в обычное время**
