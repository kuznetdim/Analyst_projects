Логика решения тут следующая:
- Мы берем "типовую квартиру" (для этого мы берем самые важные параметры, находим для них границы выбросов и отсекаем выбросы);
- Смотрим корреляцию по всем параметрам с ценами (общей и за метр), если есть что-то выше 0.5 - отмечаем в выводах;
- Но, корреляция Пирсона показывает **линейные** зависимости, а у нас могут быть еще и нелинейные (или связи с категоральными переменными). Их мы ищем при помощи графиков - scatter, hexbin, line, boxplot (по категориям для категоральных переменных). Т.е. мы должны взять все переменные, где не обнаружена высокая корреляция и исследовать их глазами на явные нелинейные зависимости).
- Этот анализ нужно провести сначала для типовой квартиры, потом отдельно для студий, апартаментов, свободных, потом для каждого из пространственных сегментов. Результаты по всем этим трем анализам нужно сравнить между собой.

Обработка выбросов, удаляем значения которые выходят за "усы"  
украл код у группы 2))  
['last_price','m_price','total_area','rooms','ceiling_height']

здесь lower_whisker = max(q_1 - 1.5*irq, df_column.min())  
находим максимальное из 2-х значений для нижнего уса, т к значение (q_1 - 1.5*irq) может быть меньше минимума  
я так предполагаю)

```python
import matplotlib.pyplot as plt
def filter_outliers(df, column_name, remove_outliers=False):
    print(column_name)
    df_column = df[column_name]
    q_1 = df_column.describe()['25%']
    q_3 = df_column.describe()['75%']
    irq = q_3 - q_1
    upper_whisker = q_3 + 1.5*irq
    lower_whisker = max(q_1 - 1.5*irq, df_column.min())
    
    df.boxplot(column_name,vert = False)
    plt.xlim(lower_whisker,upper_whisker)
    plt.show()
    
    print('Good values from',lower_whisker,'to',upper_whisker)
    outliers = df[(df_column < lower_whisker) | 
                  (df_column > upper_whisker) |
                  (df_column.isnull())]
    print('Outliers count',outliers.shape[0])
    
    if remove_outliers:
        df.drop(outliers.index, inplace = True)
        print('Outliers were removed. There are',len(df),'rows now')
    print('-'*100)
    
num_rows_before = data.shape[0]

columns = ['last_price','m_price','total_area','rooms','ceiling_height','cityCenters_nearest']
for column_name in columns:
    filter_outliers(data, column_name, remove_outliers=True)
```

```python
# хочу проставить категории 
def flat_status(row):
    if (row['studio'] == True) & (row['is_apartment'] == False) & (row['open_plan'] == False):
        return 'студия' 
    elif (row['studio'] == False) & (row['is_apartment'] == True) & (row['open_plan'] == False):
        return 'апарты'
    elif (row['studio'] == False) & (row['is_apartment'] == False) & (row['open_plan'] == True):
         return 'свободной планир'
    elif (row['studio'] == False) & (row['is_apartment'] == False) & (row['open_plan'] == False):
        return 'типовая'
    else:
        return 'неизвестно' 
data['flat_status'] = data.apply(flat_status, axis=1)
print(data['flat_status'].value_counts())
```

как я понял

```python
# хочу проставить категории 
def flat_status(row):
    if (row['studio'] == True) & (row['is_apartment'] == False) & (row['open_plan'] == False):
        return 'студия' 
    elif (row['studio'] == False) & (row['is_apartment'] == True) & (row['open_plan'] == False):
        return 'апарты'
    elif (row['studio'] == False) & (row['is_apartment'] == False) & (row['open_plan'] == True):
         return 'свободной планир'
    elif (row['studio'] == False) & (row['is_apartment'] == False) & (row['open_plan'] == False):
        return 'типовая'
    else:
        return 'неизвестно' 
data['flat_status'] = data.apply(flat_status, axis=1)
print(data['flat_status'].value_counts())
```

**в основном в выборке все квартиры типовые, остальные выборки очень маленькие**

```python
studio = data.query('flat_status in ("студия")')
free_plan = data.query('flat_status in ("свободной планир")')
standart = data.query('flat_status in ("типовая")')
apart =  data.query('flat_status in ("апарты")')
```

```python
plt.subplots(figsize = (15,6))
data.pivot_table(index='flat_status',values = 'days_exposition',aggfunc = 'median').plot(ax = plt.subplot(1, 2, 1))
data.boxplot('days_exposition',by = 'flat_status',ax = plt.subplot(1, 2, 2))
plt.suptitle('Зависимость кол-ва дней продажи от планировки',fontsize=18, y=1)
```

**выборка апартаментов, свободной планировки и студий очень маленькая по сравнению с типовой 
но здесь мы видим что кваритиры свободной планировки продаются быстрее студий(их выборки примерно одинаковые, но тоже маленькие)**

```python
plt.subplots(figsize = (15, 6))
report = studio.groupby('year').agg({'m_price': 'median'}).rename(columns = {'m_price': 'Студия'})
report = report.join(free_plan.groupby('year').agg({'m_price': 'median'}).rename(columns = {'m_price': 'Свободная планир'}))
report = report.join(standart.groupby('year').agg({'m_price': 'median'}).rename(columns = {'m_price': 'Типовая'}))
plot = report.plot(ax = plt.subplot(1, 2, 1))

report = studio.groupby('year').agg({'m_price': 'count'}).rename(columns = {'m_price': 'Студия'})
report = report.join(free_plan.groupby('year').agg({'m_price': 'count'}).rename(columns = {'m_price': 'Свободная планир'}))
report = report.join(standart.groupby('year').agg({'m_price': 'count'}).rename(columns = {'m_price': 'Типовая'}))
plot = report.plot(ax = plt.subplot(1, 2, 2))
```
